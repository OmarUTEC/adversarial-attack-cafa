```python
import pandas as pd

# Leer dataset
data = pd.read_csv("creditcard_2023.csv")

# Mostrar primeras filas
print(data.head())

# Informaci칩n general (muy importante)
print(data.info())

# Revisar valores faltantes
print(data.isna().sum())
```

```python
import matplotlib.pyplot as plt
```

```python
# Conteo de clases
conteo = data["Class"].value_counts()
print("\nConteo de transacciones:")
print(conteo)

# Renombrar para interpretaci칩n
conteo.index = ["No Fraude (0)", "Fraude (1)"]

# Gr치fico
plt.figure(figsize=(6,4))
conteo.plot(kind='bar', color=['skyblue','salmon'])
plt.title("Distribuci칩n de transacciones fraudulentas vs no fraudulentas")
plt.xlabel("Clase")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
